```python
#Numero primo
def eh_primo(n):
  if n < 2:  # Numeros menores que 2 não são primos
    return False
  if n == 2: # 2 é o único número primo par
    return True
  if n % 2 == 0: # Outros números pares não são primos
    return False
  # Verifica divisores ímpares a partir de 3 até a raiz quadrada de n
  for i in range(3, int(n**0.5) + 1, 2):
    if n % i == 0:
      return False
  return True
```

```python
#Inversão de string
def inverte_texto(Vitor):
  invertido = ""
  for letra in Vitor:
    invertido = letra+invertido
  return invertido
```

```python
print(f"'Vitor' invertido é: {inverte_texto('Vitor')}")
print(f"'Python' invertido é: {inverte_texto('Python')}")
print(f"'Olá Mundo' invertido é: {inverte_texto('Olá Mundo')}")
print(f"'a' invertido é: {inverte_texto('a')}")
print(f"'' (string vazia) invertido é: {inverte_texto('')}")
```

```python
def calculadora(num1, num2, operacao):
  """
  Realiza operações aritméticas básicas entre dois números.

  Args:
    num1: O primeiro número.
    num2: O segundo número.
    operacao: A operação a ser realizada ('+', '-', '*', '/').

  Returns:
    O resultado da operação ou uma mensagem de erro para divisão por zero ou operação inválida.
  """
  if operacao == '+':
    return num1 + num2
  elif operacao == '-':
    return num1 - num2
  elif operacao == '*':
    return num1 * num2
  elif operacao == '/':
    if num2 != 0:
      return num1 / num2
    else:
      return "Erro: Divisão por zero não é permitida."
  else:
    return "Erro: Operação inválida."
```

### Demonstração da Calculadora

```python
# Exemplos de uso da calculadora:

# Adição
resultado_soma = calculadora(10, 5, '+')
print(f"10 + 5 = {resultado_soma}")

# Subtração
resultado_subtracao = calculadora(10, 5, '-')
print(f"10 - 5 = {resultado_subtracao}")

# Multiplicação
resultado_multiplicacao = calculadora(10, 5, '*')
print(f"10 * 5 = {resultado_multiplicacao}")

# Divisão
resultado_divisao = calculadora(10, 2, '/')
print(f"10 / 2 = {resultado_divisao}")

# Divisão por zero
resultado_divisao_zero = calculadora(10, 0, '/')
print(f"10 / 0 = {resultado_divisao_zero}")

# Operação inválida
resultado_invalido = calculadora(10, 5, '%')
print(f"10 % 5 = {resultado_invalido}")
```
